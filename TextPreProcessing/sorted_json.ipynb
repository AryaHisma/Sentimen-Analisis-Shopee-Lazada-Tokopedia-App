```python
import json

# Memuat data dari file JSON
with open('assets/slang.json', 'r') as f:
    data = json.load(f)

# Mengurutkan dictionary berdasarkan kunci
sorted_data = dict(sorted(data.items()))

# Menyimpan data yang sudah diurutkan ke dalam file baru (jika diperlukan)
with open('assets/slang_sorted.json', 'w') as f:
    json.dump(sorted_data, f, ensure_ascii=False, indent=4)

# Menampilkan data yang sudah diurutkan
print(json.dumps(sorted_data, ensure_ascii=False, indent=4))
```
