```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
shopee_df = pd.read_csv('assets/dataset/scrapped_data_shopee.csv')
lazada_df = pd.read_csv('assets/dataset/scrapped_data_lazada.csv')
tokopedia_df = pd.read_csv('assets/dataset/scrapped_data_tokped.csv')
```

```python
shopee_df['at'] = pd.to_datetime(shopee_df['at'])
lazada_df['at'] = pd.to_datetime(lazada_df['at'])
tokopedia_df['at'] = pd.to_datetime(tokopedia_df['at'])

shopee_df['year'] = shopee_df['at'].dt.year
lazada_df['year'] = lazada_df['at'].dt.year
tokopedia_df['year'] = tokopedia_df['at'].dt.year
```

```python
shopee_df['marketplace'] = 'Shopee'
lazada_df['marketplace'] = 'Lazada'
tokopedia_df['marketplace'] = 'Tokopedia'

combined_df = pd.concat([shopee_df[['year', 'marketplace']],
                         lazada_df[['year', 'marketplace']],
                         tokopedia_df[['year', 'marketplace']]])
```

```python
fig = px.histogram(combined_df, 
                   x='year', 
                   color='marketplace', 
                   barmode='group', 
                   labels={'year':'Year', 'marketplace':'Marketplace'},
                   title='Comparison of Review Counts by Year')

fig.update_layout(xaxis_title='Year', 
                  yaxis_title='Number of Reviews', 
                  title_x=0.5, 
                  template='plotly_white')

fig.show()
```
